```python
# 1. Write a Python function to find the maximum of three numbers.

def max_num(n1,n2,n3):
    if n1 > n2 and n1 > n3:
        print(f'{n1} is maximum')
    elif n2 > n1 and n2 > n3:
        print(f'{n2} is max num')
    else:
        print(f'{n3} is max')



max_num(10,9,8)
    
```

```python
'''2. Write a Python function to sum all the numbers in a list.

	Sample List : (8, 2, 3, 0, 7)
	Expected Output : 20'''
from functools import reduce
def Total_Sum(l1):
    sum = 0
    for i in l1:
        sum = sum + i
    print(sum)


    total = reduce(lambda a,b:a+b,l1)
    print(total)

l1 = [8, 2, 3, 0, 7]
Total_Sum(l1)
```

```python
'''3. Write a Python function to multiply all the numbers in a list.

	Sample List : (8, 2, 3, -1, 7)
	Expected Output : -336'''
from functools import reduce
def Product(l1):
    mul = 1
    for i in l1:
        mul *= i
    print(mul)

    prod = reduce(lambda x,y:x*y,l1)
    print(prod)

l1 = [8, 2, 3, -1, 7]
Product(l1)
```

```python
'''4. Write a Python function to reverse a string. 

	Sample String : "1234abcd"
	Expected Output : "dcba4321"'''

def reverse_string(s1):
    rev = ' '
    for ch in s1:
        rev = ch + rev
    print(rev)

    
    return s1[::-1]

    

reverse_string("1234abcd")
```

```python
# 5. Write a Python function to check whether a number falls within a given range.

def check_range(n,start,end):
    
    if start <= n  and n <= end:
        print(f'{n} falls within a given range')
    else:
        print('Not in a range')

check_range(4,1,10)
```
