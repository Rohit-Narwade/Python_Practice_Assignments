```python
# 1. Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.

def div(a,b):
    try:
        return a / b
    except ZeroDivisionError:
        print('cant divide by zero')


div(10,0)    
        


```

```python
'''2. Write a Python program that prompts the user to input an integer and raises a ValueError exception 
if the input is not a valid integer.'''

def get_integer():
    input1 = input('Enter any interger: ')
    try:
        value = int(input1)
        return value
    except ValueError:
        raise ValueError('U have entered wrong value type')
try:
    num = get_integer()
    print('U entered:',num)
except ValueError as e:
    print(e)

    
```

```python
# 3. Write a Python program that opens a file and handles a FileNotFoundError exception if the file does not exist.

def read_file(filename):
    try:
        with open(filename,'r') as f:
            content = f.read()
            return content
    except FileNotFoundError:
        print('File Not Found')

data = read_file('myfile.txt')
print(data)

```

```python
# 4. Write a Python program that prompts the user to input two numbers 
# and raises a TypeError exception if the inputs are not numerical

def get_numbers():
    num1 = input('Enter 1st num: ')
    num2 = input('Enter 2nd num: ')

    try:
        value1 = float(num1)
        value2 = float(num2)
        return value1,value2
    except ValueError:
        raise TypeError('Entered wrong tyoe of data')

try:
    x,y = get_numbers()
    print('entered nos are:',x,y)
except TypeError as e:
    print(e)

```

```python
# 5. Write a Python program that opens a file and handles a PermissionError exception if there is a permission issue.
def read_file(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except PermissionError as e:
        return f"Error: {e}"

content  = read_file('myfile.txt')
print(content)
```

```python
# 6. Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range.

def get_index(l1,ind):
    try:
        result = l1[ind]
        return result
    except IndexError:
        print('Index out of range')

l1 = [1,2,3]
r = get_index(l1,4)
print(r)
```

```python
# 7. Write a Python program that executes division and handles an ArithmeticError exception if there is an arithmetic error.

def divide(a, b):
    try:
        return a / b
    except ArithmeticError:
        print("Arithmetic error occurred")


result = divide(10, 0)
print(result)

```
