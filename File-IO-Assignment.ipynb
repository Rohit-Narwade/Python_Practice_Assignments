```python
# 7. Write a python program to find the longest words.

def longest_word_finder(file):
    with open(file,'r') as f:
        words = f.read().split()
        max_length = 0
        result = []
        
        for word in words:
            if len(word) >max_length:
                max_length = len(word)
                result = [word]
            elif len(word) == max_length:
                result.append(word)
        
        return result
longest_word_finder('myfile.txt')

def longest_words(file_path):
    with open(file_path) as f:
        words = f.read().split()

    max_len = max(len(word) for word in words)
    return [word for word in words if len(word) == max_len]

longest_words('myfile.txt')
```

```python
# 9. Write a Python program to count the frequency of words in a file.

def freqency_counter(file):
    with open(file,'r') as f:
              words = f.read().split()
    freq = {}

    for word in words:
        if word in freq:
            freq[word] += 1
        else: freq[word] = 1
    return freq


freqency_counter('myfile.txt')
        
```
